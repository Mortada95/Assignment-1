```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv("Ruter_data.csv", delimiter = ";")
data.head()
```

```python
f1 = data.Dato # Vi vil at datoene på datasettet skal sorteres riktig fra minst til størst. 
f2 = data.Kjøretøy_Kapasitet # vil sortere fra lavest kapasitet til størst
f3 = data.Passasjerer_Ombord # vil sortere fra minst passasjere til størst.
```

```python
data['Dato'] = data['Dato'].astype('datetime64[ns]')
data.sort_values(by='Dato')
```

```python
data.sort_values(by='Kjøretøy_Kapasitet')
```

```python
data.sort_values(by='Passasjerer_Ombord')
```

```python
#Want to show what the usual number of people on the bus is.
plt.figure(figsize=[10,5])
plt.hist(data.Passasjerer_Ombord, bins=10, rwidth=0.8)
plt.xlabel('Passasjerer Ombord')
plt.ylabel('Antall Bussturer')
plt.show()
```

```python
#Want to see what the normal capacity of the buses is
plt.figure(figsize=[10,5])
plt.hist(data.Kjøretøy_Kapasitet, bins=10, rwidth=0.8)
plt.xlabel('Kjøretøy kapasitet')
plt.ylabel('Antall Bussturer')
plt.show()
```

```python
#Want to show what the most used bus line is

plt.figure(figsize=[10,5])
plt.hist(data.Linjenavn[:100], bins=10, rwidth=0.8)
plt.xlabel('Linjenavn')
plt.ylabel('Antall')
plt.xticks(rotation=90)
plt.show()
```

```python
#want to show the most normal province

plt.figure(figsize=[10,5])
plt.hist(data.Fylke, bins=10, rwidth=0.8)
plt.xlabel('Fylke')
plt.ylabel('Antall')
plt.xticks(rotation=90)
plt.show()
```

```python
f6 = data.Fylke
Viken = 0
Oslo = 0

for x in range(6000):
    if data.Fylke[x]=='Viken':
        Viken = Viken+1
    else:
        Oslo = Oslo+1

new_f6 = Viken
print("Viken is the most normal province with " + str(new_f6) + "busses. Removing elements " + str(Oslo) + "From the dataframe")
```

```python
# I wnat to show what the most normal "kommune" is

plt.figure(figsize=[10,5])
plt.hist(data.Kommune[:20], bins=10, rwidth=0.8)
plt.xlabel('Kommune')
plt.ylabel('Antall')
plt.xticks(rotation=90)
plt.show()
```

```python
f1 = data.Kommune
Bærum = 0
Bjerke = 0
St_Hanshaugen = 0
Indre_Ostfold = 0
Sentrum = 0
Ullensaker = 0
Eidsvoll = 0


for x in range(6000):
    if data.Kommune[x]=='Bærum':
        Bærum = Bærum+1
    elif data.Kommune[x]=='Bjerke':
        Bjerke = Bjerke+1
    elif data.Kommune[x]=='St_Hanshaugen':
        St_Hanshaugen = St_Hanshaugen+1
    elif data.Kommune[x]=='Indre_Ostfold':
        Indre_Ostfold = Indre_Ostfold+1
    elif data.Kommune[x]=='Sentrum':
        Sentrum = Sentrum+1
    elif data.Kommune[x]=='Ullensaker':
        Ullensaker = Ullensaker+1
    else:
        Eidsvoll = Eidsvoll+1

        
Kommune = np.array([Bærum,Bjerke,St_Hanshaugen,Indre_Ostfold,Sentrum,Ullensaker,Eidsvoll])
print(Kommune)
print("From this array we can se that Eidsvoll is the one fylke with most traffic")

        
```
